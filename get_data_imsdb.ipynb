```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import pickle
import time
import random
import re
from copy import deepcopy
```

Get movie URL from IMSDB

```python
script_url = {}
url = 'http://www.imsdb.com/alphabetical/' + '0'
response = requests.get(url)
page = response.text
soup = BeautifulSoup(page, 'lxml')
ps = soup.find_all('p')
for p in ps:
    link = p.find('a')
    script_url[link['title']] = link['href']
```

Scrape the movie information and the scripts

```python
def get_info(soup, name):
    movie_info=dict(title=name)
    try:
        rate_regex = re.compile(r'\d+\.\d+ out of 10')
        rate_raw = soup.find(text=rate_regex)
        m = re.search(r'(\d+\.\d+) out of 10', rate_raw)
        movie_info['rating'] = m.group(1)
    except:
        print('cannot get rating for movie %s' % name)
    try:
        genre = []
        table = soup.find('table', {'class': 'script-details'})
        all_links = table.find_all('a')
        for l in all_links:
            if l['href'].find('genre') != -1:
                genre.append(l['href'].strip('/genre/'))
        movie_info['genre'] = genre
    except:
        print('cannot get genre for movie %s' % name )
    try:
        release = table.find('b', text='Movie Release Date')
        release_time = release.next_sibling.strip(' :')
        movie_info['release_time'] = release_time
    except:
        print('cannot get release time for movie %s' % name )
    try:
        month, year = release_time.split()
        movie_info['month'] = month
        movie_info['year'] = year
    except:
        print('cannot parse time to month and year for movie %s' % name)
    try:
        movie_info['script_url'] = release.find_next('a')['href']
    except: 
        try:
            read_regex = re.compile(r'^Read')
            movie_info['script_url'] = table.find('a', text=read_regex)['href']
        except:
            print('cannot get url for movie %s' % name )
    return movie_info

    
```

```python
#get movie info and script url
movie_meta = []
i = 0
for name, url in script_url.items():
    full_url = 'http://www.imsdb.com' + url
    response = requests.get(full_url)
    page = response.text
    soup = BeautifulSoup(page, 'lxml')
    movie_meta.append(get_info(soup, name))
    i += 1
    if i%7 == 0:
        time.sleep(random.random()*3)
    if i%100 == 0:
        print('scraping', i)
```

```python
#get script text
def get_script(url):
    full_url = 'http://www.imsdb.com' + url
    response = requests.get(full_url)
    page = response.text
    soup = BeautifulSoup(page, 'lxml')
    try: 
        script_text = soup.find('pre').text
        return script_text
    except:
        raise ValueError('cannot get the script from url', full_url)
```

```python
i = 0 
unsuccess={}
for movie in movie_meta:
    try:
        movie['scripts'] = get_script(movie['script_url'])
    except:
        print('cannot get script for ', movie['title'])
        if 'script_url' in movie:
            unsuccess[movie['title']] = movie['script_url']
    i += 1
    if i%7 == 0:
        time.sleep(random.random()*3)
    if i%200 == 0:
        print('working on ', i)
        filename = 'data/movie_meta_'+str(i//200)+'.pkl'
        with open(filename, 'wb') as f:
            pickle.dump(movie_meta, f)
```

```python
#remove entries without url or scripts
movie_meta_f = [movie for movie in movie_meta if 'script_url' in movie and 'scripts' in movie]
```
