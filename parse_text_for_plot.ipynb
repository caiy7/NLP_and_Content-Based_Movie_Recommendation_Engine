```python
%matplotlib inline
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from nltk.corpus import stopwords
import string
```

```python
with open('data/movie_meta_final_clean.pkl', 'rb') as f:
    movie_meta = pickle.load(f)
```

```python
def clean_text(text):
    text = text.strip()
    b_tags = re.findall('<b>', text)
    if len(b_tags)<6:
        text = re.sub('<b>[\w\W]*<\/b>', '', text)

    text = re.sub('.*DRAFT.*', '', text)
    text = re.sub('Sc\s+\d.*', '', text)
    text = re.sub('É|é', 'e', text) 
    text = re.sub('á|ä', 'a', text)
    text = re.sub('.*\d+[_\.\-\/]\d+[_\.\-\/]\d+.*', '', text, flags=re.IGNORECASE) #remove date
    text = re.sub('.*\d+ \d+.*', '', text, flags=re.IGNORECASE)
    text = re.sub('.*revision.*', '', text, flags=re.IGNORECASE)
    text = re.sub('.*(int |ext |int\.|ext\.|close on|angle|high wide|exterior|interior|full shot|dissolve to|open on|shooting script|c\.u\.|long shot|med\. shot).*',
              '\n', text, flags=re.IGNORECASE)
    remove_list = ['cut to:', 'continued:', 
              'close-up:', 'med. shot:', 'close shot:', 'fade in', 'day', 'continued',
                '(v.o.)', 'scene deleted', "cont'd", 'beat.', 'contd',
              '(continuing)', '(v/o)']
    for word in remove_list:
        text = text.lower().replace(word, '')
    return text
```

```python
movie_meta[:] = [movie for movie in movie_meta if movie['title'] not in movie_to_remove]
```

```python
for movie in movie_meta:
    print(movie['title'])
    movie['clean_script'] = clean_text(movie['scripts'])
```
